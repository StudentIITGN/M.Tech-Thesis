```python
from obspy import UTCDateTime
from obspy.clients.fdsn import Client
import csv
from obspy.clients.fdsn.header import FDSNNoDataException
```

```python
# importing all neccesary libaries to make plot

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
client=Client("IRIS")     #Incorporated Reasearch Institutions for seismology

#provided a sourse from where to download data
#defining a timeperiod

time_periods = [
    (UTCDateTime("2019-01-01"), UTCDateTime("2024-01-02")),
    (UTCDateTime("2015-01-01"), UTCDateTime("2019-01-02")),
    (UTCDateTime("2010-01-01"), UTCDateTime("2015-01-02")),
    (UTCDateTime("2009-01-01"), UTCDateTime("2010-01-02")),
    (UTCDateTime("2008-01-01"), UTCDateTime("2009-01-02")),
    (UTCDateTime("2007-01-01"), UTCDateTime("2008-01-02")),
    (UTCDateTime("2006-01-01"), UTCDateTime("2007-01-02")),
    (UTCDateTime("2005-01-01"), UTCDateTime("2006-01-02")),
    (UTCDateTime("2000-01-01"), UTCDateTime("2005-01-02")),
    (UTCDateTime("1995-01-01"), UTCDateTime("2000-01-02")),
    (UTCDateTime("1990-01-01"), UTCDateTime("1995-01-02")),
    (UTCDateTime("1985-01-01"), UTCDateTime("1990-01-02")),
    (UTCDateTime("1980-01-01"), UTCDateTime("1985-01-02")),
    (UTCDateTime("1975-01-01"), UTCDateTime("1980-01-02")),
    (UTCDateTime("1970-01-01"), UTCDateTime("1975-01-02")),
    (UTCDateTime("1965-01-01"), UTCDateTime("1970-01-02")),
    (UTCDateTime("1960-01-01"), UTCDateTime("1965-01-02")),
    (UTCDateTime("1955-01-01"), UTCDateTime("1960-01-02")),
    (UTCDateTime("1950-01-01"), UTCDateTime("1955-01-02")),
    (UTCDateTime("1945-01-01"), UTCDateTime("1950-01-02")),
    (UTCDateTime("1940-01-01"), UTCDateTime("1945-01-02")),
    (UTCDateTime("1935-01-01"), UTCDateTime("1940-01-02")),
    (UTCDateTime("1930-01-01"), UTCDateTime("1935-01-02")),
    (UTCDateTime("1925-01-01"), UTCDateTime("1930-01-02")),
    (UTCDateTime("1920-01-01"), UTCDateTime("1925-01-02")),
    (UTCDateTime("1915-01-01"), UTCDateTime("1920-01-02")),
    (UTCDateTime("1910-01-01"), UTCDateTime("1915-01-02")),
    (UTCDateTime("1905-01-01"), UTCDateTime("1910-01-02")),
    (UTCDateTime("1900 -01-01"), UTCDateTime("1905-01-02")),
    
]




#provide a geographical bounds for the region
minlatitude =8.781306565948524
maxlatitude =  17.74606593415912                      
minlongitude =81.47453848358678
maxlongitude =101.41627183447663


#provide file path where data will get stored

filename =r"C:\Users\Dakshata\OneDrive - iitgn.ac.in\Thesis\Thesis\All_SEA\SEA2\notebooks\Seismicdata_SEA2.csv"
#r is a raw string to deal with escape character


with open(filename,mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['Time','Latitude','Longitude','Depth (km)','Magnitude'])
    
    for start,end in time_periods:
        try:
            cat=client.get_events(starttime=start,endtime=end,
                                 minmagnitude=0.0,minlatitude=minlatitude,maxlatitude=maxlatitude,
                                 minlongitude=minlongitude,maxlongitude=maxlongitude)
            
            for event in cat:
                time=event.origins[0].time
                latitude=event.origins[0].latitude
                longitude=event.origins[0].longitude
                depth=event.origins[0].depth
                if depth is not None:
                    depth/=1000.0
                    #converting depth to kilometer 
                magnitude=event.magnitudes[0].mag
                writer.writerow([time,latitude,longitude,depth,magnitude])
            print("Downloaded data is saved to:",filename)
            
        except FDSNNoDataException as e:
            print("No seismic data available for the criteria you specified:",e)
        except Exception as e:
            print("An error occured:",e)
```
